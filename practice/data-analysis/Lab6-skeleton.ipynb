````python
import numpy as np
import pandas as pd


ratings_path = 'ratings.csv'
movies_path = 'movies.csv'
ratings = pd.read_csv(ratings_path)
movies = pd.read_csv(movies_path)
````

(1) 'ratings.csv'파일을 읽어와 datafame을 인자로 받고, 각 영화별 평점 평균을 구하는 함수 get_ratings_pandas()를 구현하세요.    
(단, pandas를 사용하여 구현하세요.)   
   

````python
print(ratings.describe())
print(ratings.head())
# ratings[['movieId', 'rating']]
def get_ratings_pandas(df):
    return ratings[['movieId', 'rating']].groupby('movieId').mean()
print(get_ratings_pandas(ratings))
````

(2) ratings에서 movie의 평균 평점이 4이상인 영화의 비율을 구하세요.   
(단, 영화 당 평점 데이터가 5개 이하인 경우는 제외합니다.)

````python
ratings_by_movieId = get_ratings_pandas(ratings)
print(ratings_by_movieId)
print(ratings_by_movieId[(ratings_by_movieId['rating'] > 4)])
# get_ratings_pandas(ratings)['rating'] > 4
````

(3) ratings와 movies의 두개의 데이터프레임을 이용해 title별 영화 평점 평균을 구하세요.

````python
movies_ratings = pd.merge(movies, ratings, on='movieId')
print(movies_ratings)
print(movies_ratings.groupby('title').mean()['rating'])
````

(1) `vaccine_seoul.csv` 파일을 읽어 데이터프레임 `df_vaccine`를 만들어라. 
 

(2) 일별 확진자의 수와 일별 2차접종률(%)을 담은 데이터프레임 `df`을 만드세요.
  
* 데이터프레임`df`에는 확진자수와 2차 접종률(%) 자료가 모두 있는 날짜만 저장되면 된다.    
* 날짜와 그날의 확진자의 수를 담은 데이터프레임 `df1`을 만들어라.   
참고) `df.reset_index()`를 사용하면, 데이터프레임`df`의 인덱스를 재설정한다.  단, 이때 `df`를 변경하지 않으니, 변경하고 싶다면 변수에 저장하여 사용.
* 날짜(접종일)과 그날의 2차접종률(%)을 담은 데이터프레임 `df2`를 만들어라.   
 
참고) 각 열의 이름은 `df.columns`를 지정하여 변경 가능. ex) `df.columns = ['a', 'b']`  
참고) `pd.merge()` 이용 

````python
df = pd.read_csv("vaccine_seoul.csv")
df_covid = pd.read_csv("covid19_seoul.csv", encoding="EUC-KR")
````

````python
print(type(df_covid.groupby('확진일').count()))
df_covid_date = df_covid.groupby('확진일').size().to_frame()
df_covid_date.reset_index(inplace=True)
df_covid_date.columns = ['날짜', '확진자수']
df_covid_date.head()
````

````python
# df_covid.info()
# print(df_covid.head())
# df.info()
# print(df.head())
# df['2차접종률(%)'].isna().sum()
# df['2차접종률(%)'].isna().sum()
df_vac_date = df[['접종일', '2차접종률(%)']]
df_vac_date.columns = ['날짜', '2차접종률(%)']
df_vac_date.head()

pd.merge(df_covid_date, df_vac_date, on='날짜', how='inner')
pd.merge(df_covid_date, df_vac_date, on='날짜')
````
