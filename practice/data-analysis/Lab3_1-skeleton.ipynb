딕셔너리에서 value가 None인 원소 제거
```
dict1 = {'a': 'apple', 'b': 'banana', 'c':None}

dict1 = {key:value for (key, value) in dict1.items() if value is not None}
print(dict1)

# Returns: {'a': 'apple', 'b': 'banana'}
```

딕셔너리에서 필터를 사용하여 값 제거 (dictionary comprehension 사용)
```
heights = {'John': 175, 'Jane': 150, 'Jim': 155, 'Matt': 170}

tall = {key:value for (key, value) in heights.items() if value >= 170}

print(tall)
# Returns: {'John': 175, 'Matt': 170}
```

**문제2**   
주어진 텍스트 파일 (log)은 서버 접속이 차단된 ip입니다. 접속 ip의 상위 10개와 그 횟수를 구해보세요.   

결과) 
```
# 접속 상위 ip, 횟수
[('113.180.164.102', 280),    
('110.78.138.88', 280),    
('183.88.144.150', 280),    
('113.161.160.16', 280), 
('171.100.9.130', 280),    
('36.92.160.250', 280),    
('1.55.211.148', 278),    
('103.230.155.58', 242),    
('113.53.129.247', 153),    
('83.122.161.57', 99)]
```

````python
with open("log") as f:
    d = dict()
    for l in f:
        words = l.split()
        if len(words) > 2:
            ip = words[2]
            if len(ip.split('.')) == 4:
                d[ip] = d.get(ip, 0) + 1
    ll = [(k, v) for v, k in sorted([(v, k) for k, v in d.items()], reverse=True)[:10]]

    print(sorted(d.items(), key=lambda item: item[1], reverse=True)[0:10])
    print(ll)
````

**문제3**   
다양한 Phone number의 형식을 통일된 형식으로 출력해보세요.   
(아래 예시를 처리할 수 있도록 만들기)

결과) 
``` 
#Before
010-2220-3434  
01012345678  
010-12345675  
공일공-2220-3434
공일공2220오칠삼육
```

```
#After
01022203434
01012345678
01012345675
01022203434
01022205736
```


````python
def f(phoneNum):
    numDict = {"공":"0", "일":"1", "이":"2", "삼":"3", "사":"4", "오":"5", "육":"6", "칠":"7", "팔":"8", "구":"9"} 
    return "".join([ numDict[num] if num in numDict else num for num in phoneNum if (num >= "0" and num <= "9") or num in numDict ])

f("공일공-2220-3434")
````
